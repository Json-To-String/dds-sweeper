```python
import math
math.floor(1/0.016)
# ## python way
# def get_continued_fraction(number, num_terms = 3):
#     frac_list = [math.floor(number)]
#     for k in range(num_terms):
#         f = number - frac_list[k]
#         print(f)
#         number = 1/f
#         print(number)
#         frac_list.append(math.floor(number))
#     return(frac_list)

## c like way
num_terms = 4
frac_list = [None]*num_terms

def get_continued_fraction_C(number, num_terms=3):

    frac_list[0] = math.floor(number)
    for i in range(num_terms):
        diff = number - frac_list[i]
        
        number = 1/diff
        frac_list[i + 1] = math.floor(number)
    
    print(f"Now frac_list is {frac_list}")

# print(get_continued_fraction(number = 2.3184))

print(get_continued_fraction_C(number = 2.3184))

def list_to_frac(inputlist):
    n = len(inputlist)
    if n==0:
        return(0, 0)
    elif n==1:
        return(inputlist[0], 0)
    else:
        p0 = inputlist[0]
        q0 = 1
        p1 = inputlist[1] * inputlist[0] + 1
        q1 = inputlist[1]
        for k in range(3, n + 1):
            p = inputlist[k - 1] * p1 + p0
            q = inputlist[k - 1] * q1 + q0
            p0 = p1
            p1 = p
            q0 = q1
            q1 = q
        return(p1, q1, p0, q0)
print(list_to_frac([2, 3, 1, 4]))
print(list_to_frac([2, 3, 7, 9]))

def approx_double(f, p_max, q_max, num_iters = 6):
    eps = 0.001

    a0 = math.floor(f)
    r_old = f - a0
    q_double_old = 0
    p_double_old = 1
    p_old = a0
    q_old = 1

    a_new, p_new, q_new = None, None, None
    r_new = None, None

    for i in range(1, num_iters):
        print(i)
        check = f - (p_old / q_old)
        if (abs(check) < eps):
            print('number close within eps')
            break

        a_new = math.floor(1/r_old)
        p_new = a_new * p_old + p_double_old
        q_new = a_new * q_old + q_double_old

        if (p_new > p_max or q_new > q_max):
            print('either num or denom too big')
            break
        r_old = 1/r_old - a_new

        p_double_old = p_old
        p_old = p_new
        q_double_old = q_old
        q_old = q_new

    return(p_old, q_old, i)

x = approx_double(2.314075007934069, 2**32 - 1, 256, num_iters=100)
print(x)
print(x[0] / x[1])
```
